```
# keunggulan python library . . . library skimage
from skimage import  io
import matplotlib.pyplot as plt

img = io.imread('https://www.cdc.gov/dpdx/malaria/images/1/Pf_rings_thinA.jpg')
fig,ax = plt.subplots(2,2,figsize=(15,10))
ax[1,1].imshow(img)
ax[0,1].imshow(img)
plt.show
```
